```python
def get_prodict(x:int, y:int) -> int:
    return x*y


get_prodict(x=2, y=4)
```

```python
get_prodict(x=2, y="4")
```

```python
from ensure import ensure_annotations

@ensure_annotations
def get_prodict(x:int, y:int) -> int:
    return x*y
```

```python
print(get_prodict(x=2, y=4))
```

```python
print(get_prodict(x=2, y="4"))
```
