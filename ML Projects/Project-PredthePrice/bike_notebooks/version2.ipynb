2. remove km from the end of kms_driven

```python
def getKMdriven(kms):
    try:
        if kms.startswith('Mileage'):
            km = int(kms.split()[1])
        else:
            km = int(kms.split()[0])
        return km
    except ValueError:
        return kms.split()[1]

df['km_driven'] = df['kms_driven'].apply(getKMdriven)
```

```python
for index, row in df.iterrows():
    if type(row['km_driven']) != int:
        print((index, row['km_driven']))
```

```python
# resolving dirty and invalid values
def getKMdriven(kms):
    try:
        if kms.startswith('Mileage'):
            km = int(kms.split()[1])
        elif kms.endswith('Kmpl'):
            km = int(kms[:2])
        else:
            km = int(kms.split()[0])
        return km
    except ValueError:
        km = kms.split()[1]
        if km == 'Km':
            return km
        else:
            return int(km[:2])

df['km_driven'] = df['kms_driven'].apply(getKMdriven)
```

```python
for index, row in df.iterrows():
    if row['km_driven'] == 'Km':
        df.drop(index, inplace=True)
```
